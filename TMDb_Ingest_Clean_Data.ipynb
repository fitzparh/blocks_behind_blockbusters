## Transforming JSON columns

```python
#let's use a some for loops to identify all the JSON files that need to flattened into subsequent columns
json_cols = []
for c, i in df.iteritems():
    if type(df[c].loc[0]) == str:
        if "[{" in (df.loc[0,c]):
            json_cols.append(c)

#let's create a function to split those columns using '|' delimiter            
def pipe_flatten_names(col):
    return '|'.join(x['name'] for x in col[:3])

#let's use the 'json.loads' method + flatten name function, taking only the first 3 values
for col in json_cols:
    df[col] = df[col].apply(json.loads)
    df[col+"_first3"] = df[col].apply(pipe_flatten_names)
```

## Further Feature Creation

```python
#let's create a function to get the director from crew column
def grab_director(crew_col):
    directors = [x['name'] for x in crew_col if x['job'] == 'Director']
    if len(directors) > 0:
        return directors[0]
    else:
        return np.nan

#apply grab_director function
df['director'] = df.crew.apply(grab_director)

#split the production companies, taking the first value
df['film_country'] = df.production_countries_first3.str.split('|')
df['film_country'] = [x[0] for x in df['film_country']]

#split the cast column, take the first 3 values
df['actor_split'] = df.cast_first3.str.split('|')
df['actor_1'] = [x[0] for x in df['actor_split']]
df['actor_2'] = [x[1] if len(x)>=2 else x[0] for x in df['actor_split']]
df['actor_3'] = [x[2] if len(x)>=3 else x[0] for x in df['actor_split']]

#split the cast column, take the first 3 values
df['p_comp_split'] = df.production_companies_first3.str.split('|')
df['p_comp_1'] = [x[0] for x in df['p_comp_split']]
df['p_comp_2'] = [x[1] if len(x)>=2 else x[0] for x in df['p_comp_split']]
df['p_comp_3'] = [x[2] if len(x)>=3 else x[0] for x in df['p_comp_split']]

#create a binary label for to mark whether a film is spoken at all in english i.e. 0 or is exclusively non-english = 1
df['No_English'] = [0 if 'English' in ele else 1 for ele in df['spoken_languages_first3']]

```
